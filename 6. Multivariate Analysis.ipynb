# Multivariate Analysis

- Multivariate analysis is based on the statistical principle of multivariate statistics, which involves observation and analysis of more than one statistical outcome variable at a time.

- Many decisions are based on univariate analysis, but only multivariate analysis reveals relationships that help you detect problems that are not obvious by looking at the variables individually.

# Interval(0.16-10L)

```python
c=df[(df['price(L)']>=0.16)&(df['price(L)']<=10.00)]
d6_sort=c.sort_values(by='price(L)',axis=0,ascending=False)
```

```python
def multi_scatter(col1,col2,col3,x,y):
    c=df[(df['price(L)']>=x)&(df['price(L)']<=y)]
    d6_sort=c.sort_values(by='price(L)',axis=0,ascending=False)
    sns.set_style("whitegrid")
    ax1=sns.FacetGrid(d6_sort.head(30),hue=col3,height=5,aspect=2).map(plt.scatter,col1,col2,marker='+').set(yscale='log').add_legend()
    ax1.set_xticklabels(rotation=60)
    d6_sort=c.sort_values(by='price(L)',axis=0,ascending=True)
    sns.set_style("whitegrid")
    ax2=sns.FacetGrid(d6_sort.head(30),hue=col3,height=5,aspect=2).map(plt.scatter,col1,col2,marker='+').set(yscale='log').add_legend()
    ax2.set_xticklabels(rotation=60)
    plt.show()
    
```

```python
multi_scatter('location','price(L)','building_status',0.16,10)
```

#### Observation:
###### Graph 1:
   - For the price range of 16K to 10L from high to low price for **top 30**
   - Mostly,highest price  of 10L is there in all locations but new buildings are more from Yadagirigutta to Shadnagar.
###### Graph 2:
   - For the price range of 16K to 10L from high to low price for **bottom 30**
   - Moslty,lowest price  of 1.75L is there for new building at Kadthal,Shadnagar,Vikarabad.

```python
multi_scatter('location','price(L)','title',0.16,10)
```

#### Observation:
###### Graph 1:
   - For the price range of 16K to 10L from high to low price for **top 30**
   - Mostly highest price  of 10L is there in all loacations
   - From kadthal to Srisailam Highway there are more no of Residential plots
###### Graph 2:
   - For the price range of 16K to 10L from high to low price for **bottom 30**
   - Moslty,lowest price  of 1.75L is there for  Residential plots in Kadthal Srisailam and Thandur.
